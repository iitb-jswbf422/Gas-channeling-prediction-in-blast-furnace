```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
from matplotlib.path import Path

# Define the vertices of the blast furnace outline for half view
vertices = np.array([
    [0.0, 0.0],
    [0.4, 0.0],
    [0.4, 0.2],
    [0.5, 0.4],
    [0.5, 0.6],
    [0.3, 1.4],
    [0.3, 1.8],
    [0.0, 1.8],
    [0.0, 0.0]
])


# Create a meshgrid within the triangle
x = np.linspace(0, 1.5, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x, y)

# Define a path for the triangle
path = Path(vertices)

# Create a mask for points outside the triangle
mask = path.contains_points(np.vstack((X.flatten(), Y.flatten())).T)
mask = mask.reshape(X.shape)

# Function for temperature variation
def temperature(x, y):
    return x**2 + y**2  # A simple function for demonstration

Z = np.ones_like(X) * np.min(temperature(X, Y))  # Set temperature outside triangle to minimum value
Z[mask] = temperature(X[mask], Y[mask])  # Set temperature inside triangle

from matplotlib.colors import LinearSegmentedColormap
colors = [(1, 1, 1), (1, 1, 0), (1, 0, 0)]  # White to yellow to red
cmap_name = 'custom_temperature'
custom_cmap = LinearSegmentedColormap.from_list(cmap_name, colors)

# Plot the contour
plt.figure()
plt.contourf(X, Y, Z, levels=20, cmap=custom_cmap)  

# Plot the vertices
plt.plot(vertices[:,0], vertices[:,1], 'ro')  # 'ro' means red circles
plt.gca().add_patch(PathPatch(path, fill=None, edgecolor='r'))

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Temperature Variation within Triangle')
plt.colorbar(label='Temperature')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
from matplotlib.patches import PathPatch
from matplotlib.path import Path

# Define the blast furnace outline vertices for half view
vertices = np.array([
    [0.0, 0.0],
    [0.4, 0.0],
    [0.4, 0.2],
    [0.5, 0.4],
    [0.5, 0.6],
    [0.3, 1.6],
    [0.3, 1.8],
    [0.0, 1.8],
    [0.0, 0.0]
])

# Define Y-axis intervals corresponding to ROW08 and ROW09
row_intervals = {
    "ROW08": (0.5, 0.6),
    "ROW09": (0.6, 0.7)
}

# Define colors for the colormap
colors = ['green', 'yellow', 'orange', 'red']

# Assuming 's', 'L', 'df1' are defined earlier
for n in s.columns:
    # Extract data for Q1 and Q2 corresponding to ROW08 and ROW09
    data = {
        "Q1": [df1[L[n]][3], df1[L[n]][7]],
        "Q2": [df1[L[n]][4], df1[L[n]][8]]
    }

    # Create a DataFrame with the data and specified row indices
    df5 = pd.DataFrame(data, index=["ROW08", "ROW09"])

    # Create the figure and axes
    plt.figure(figsize=(20, 6))
    ax = plt.gca()

    # Plot the blast furnace outline
    path = Path(vertices)
    patch = PathPatch(path, edgecolor='r', facecolor='none')
    ax.add_patch(patch)

    # Set limits for x and y axes
    ax.set_xlim(0, 0.6)
    ax.set_ylim(0, 2)

    # Plot the contour within the blast furnace outline and row intervals
    for row_label, (y_min, y_max) in row_intervals.items():
        if row_label in df5.index:
            # Define the y-values within the specified interval
            y_vals = np.linspace(y_min, y_max, 100)
            
            # Create meshgrid for contour plot
            X, Y = np.meshgrid(["Q1", "Q2"], y_vals)
            Z = np.array(df5.loc[row_label]).reshape(1, -1)
            
            # Plot contour within the defined region
            ax.contourf(X, Y, Z, cmap=mcolors.ListedColormap(colors), alpha=1, levels=[40, 60, 75, 120])

    # Set labels and title
    ax.set_xlabel('Quadrant Number')
    ax.set_ylabel('Height Level')
    ax.set_title("Channeling {} - Date {}".format(df1[L[n]][2], df1[L[n]][1]), fontsize=15)

    # Add colorbar
    cb = plt.colorbar()
    cb.set_label('Pressure')

    # Show the plot
    plt.show()
    plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
from matplotlib.path import Path

# Define the vertices of the blast furnace outline for half view
vertices = np.array([
    [0.0, 0.0],
    [0.4, 0.0],
    [0.4, 0.2],
    [0.5, 0.4],
    [0.5, 0.6],
    [0.3, 1.6],
    [0.3, 1.8],
    [0.0, 1.8],
    [0.0, 0.0]
])

# Limit the x-axis to 0.6
x_limit = 0.6

# Create a meshgrid within the defined range
x = np.linspace(0, x_limit, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x, y)

# Define a path for the triangle with adjusted vertices within x-axis limit
adjusted_vertices = vertices.copy()
adjusted_vertices[:, 0] = np.clip(adjusted_vertices[:, 0], None, x_limit)  # Clip x values to x_limit
path = Path(adjusted_vertices)

# Create a mask for points outside the triangle
mask = path.contains_points(np.vstack((X.flatten(), Y.flatten())).T)
mask = mask.reshape(X.shape)

# Plotting the meshgrid and the mask
plt.figure(figsize=(20, 10))

# Plot the meshgrid within the specified x-axis limit
plt.imshow(mask, extent=(0, x_limit, 0, 2), origin='lower', cmap='gray')

# Plot the outline of the adjusted triangle
plt.plot(adjusted_vertices[:, 0], adjusted_vertices[:, 1], 'r-')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Furnace height')
plt.title('Blast furnace')

# Show the plot
plt.show()
```
